```python
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')
display(df.head())
display(df.dtypes)
```

```python
df['Show_Id'] = df['Show_Id'].str.replace('s', '').astype(int)
display(df.head())
```

```python
df.info()
```

```python
movies_df = df[df['Category'] == 'Movie'].copy()
tvshows_df = df[df['Category'] == 'TV Show'].copy()
```

```python
movies_df['Duration'] = movies_df['Duration'].str.replace(' min', '').astype(int)
display(movies_df.head())
```

```python
tvshows_df['Duration'] = tvshows_df['Duration'].str.replace(' Seasons', '').str.replace(' Season', '').astype(int)
display(tvshows_df.head())
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df['Cast'] = df['Cast'].fillna('Unknown Cast')
display(df.isnull().sum())
```

```python
df['Director'] = df['Director'].fillna('Unknown')
display(df.isnull().sum())
```

```python
display(df['Type'].value_counts())
```

```python
df['Country'] = df['Country'].fillna('Unknown Country')
display(df.isnull().sum())
```

```python
df.dropna(subset=['Rating'], inplace=True)
display(df.isnull().sum())
```

```python
df.dropna(subset=['Release_Date'], inplace=True)
display(df.isnull().sum())
```

```python
# Content type distribution
sns.countplot(data=df, x='Category')
plt.title('Movies vs TV Shows')
plt.show()
```

```python
# Top 10 countries with most content
top_countries = df['Country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title('Top 10 Countries with Most Netflix Content')
plt.show()
```

```python
# Trend of movies vs TV shows by year
df['release_year'] = df['Release_Date'].dt.year
content_trend = df.groupby(['release_year', 'Category']).size().unstack().fillna(0)
content_trend.plot(kind='line', figsize=(10,6), title='Movies vs TV Shows Over Time')
plt.ylabel('Number of Titles')
plt.xlabel('Release Year')
plt.show()
```

```python
# Top 10 Directors on Netflix
top_directors = df['Director'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_directors.values, y=top_directors.index, palette='viridis')
plt.title('Top 10 Directors on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Director')
plt.show()
```

```python
# Clean duration column (Movies only)
# Filter movies
movies = df[df['Category'] == 'Movie'].copy()
movies['Duration'] = movies['Duration'].str.replace(' min', '').astype(int)
avg_duration = movies['Duration'].mean()

print(f"📽️ Average Movie Duration: {avg_duration:.1f} minutes")
```

```python
top_countries = df['Country'].value_counts().head(15)

plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='coolwarm')
plt.title('Top 15 Countries Producing Netflix Content')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

```python
# Group by year and plot
content_by_year = df['Release_Date'].dt.year.value_counts().sort_index()

plt.figure(figsize=(10,5))
sns.lineplot(x=content_by_year.index, y=content_by_year.values, marker='o')
plt.title('Number of Titles Added to Netflix Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Titles Added')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=movies, x='release_year', y='Duration', alpha=0.5)
plt.title('Movie Duration vs Release Year')
plt.xlabel('Release Year')
plt.ylabel('Duration (minutes)')
plt.show()
```
